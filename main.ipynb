Imports

```python
import pandas as pd
import seaborn as sns
```

Read in CSV

```python
df = pd.read_csv('TWO_CENTURIES_OF_UM_RACES.csv')
df.head()
```

```python
df.dtypes
```

Clean up data

```python
#Step 1 show 50mi or 50km
#50km
#50mi
```

```python
df[df['Event distance/length'] == '50km']
```

Create dataframe with events in USA filtered by year 2020

```python
df2 = df[
    (df['Event distance/length'].isin(['50mi', '50km'])) &
    (df['Year of event'] == 2020) & 
    (df['Event name'].str.contains('(USA)'))
    ]
```

```python
df2.head()
```

remove (USA) from event name

```python
df2['Event name'] = df2['Event name'].str.replace('(USA)', '')
df2.head()
```

Calculate athlete age

```python
df2['Athlete age'] = df['Year of event']  - df2['Athlete year of birth'] 
df2.head()
```

Remove h from athlete performance

```python
df2['Athlete performance'] = df2['Athlete performance'].str.replace('h', '')
df2.head()
```

Drop columns: Athlete Club, Athlete Country, Athlete Year of Birth, Athlete Age Category

```python
# df2 = df2.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis=1)
df2.head()
```

Clean up null values

```python
df2.isna().sum()
```

```python
df2[df2['Athlete age'].isna()]
```

```python
df2 = df2.dropna()
```

Check for duplicate values

```python
df2[df2.duplicated()] == True
```

Reset index

```python
df2.reset_index(drop=True)
```

Fix types

```python
df2['Athlete age'] = df2['Athlete age'].astype(int)
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2.dtypes
```

Rename columns

```python
# Year of event                  int64
# Event dates                   object
# Event name                    object
# Event distance/length         object
# Event number of finishers      int64
# Athlete performance           object
# Athlete gender                object
# Athlete average speed        float64
# Athlete ID                     int64
# Athlete age                    int32
```

```python
df2 = df2.rename(columns= {
   'Year of event': 'year',
   'Event dates': 'race_day',
   'Event name': 'race_name',
   'Event distance/length': 'race_length',
   'Event number of finishers': 'race_number_of_finishers',
   'Athlete performance': 'athlete_performance',
   'Athlete gender': 'athlete_gender',
   'Athlete average speed': 'athlete_average_speed',
   'Athlete ID': 'athlete_id',
   'Athlete age': 'athlete_age', 
})
```

```python
df2.head()
```

Reorder columns

```python
df3 = df2[['race_day', 'race_name', 'race_length', 'race_number_of_finishers', 'athlete_id', 'athlete_gender', 'athlete_age', 'athlete_performance', 'athlete_average_speed', 'year']]
```
